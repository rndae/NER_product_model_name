```python
import os
import json
import gzip
import pandas as pd
```

```python
data = []
list_of_files = ["listings_0.json.gz","listings_1.json.gz","listings_2.json.gz","listings_3.json.gz","listings_4.json.gz","listings_5.json.gz","listings_6.json.gz","listings_7.json.gz","listings_8.json.gz","listings_9.json.gz","listings_a.json.gz","listings_b.json.gz","listings_c.json.gz","listings_d.json.gz","listings_e.json.gz","listings_f.json.gz"]

for file in list_of_files:
    with gzip.open('amazon-listings/' + file) as f:
        for l in f:
            data.append(json.loads(l.strip()))
    

print(len(data))

print(data[0])
```

```python
data_with_model_name = [x for x in data if 'model_number' in x and 'brand' in x]
len(data_with_model_name)
```

```python
data_with_model_name[890]
```

```python
data_english_with_models = []

for row in data_with_model_name:
    title = ""
    brand = ""
    model = ""
    for lang_tag in row['item_name']:
        if "en_" in lang_tag['language_tag']:
            title = lang_tag['value']
            break
    for lang_tag in row['brand']:
        if "en_" in lang_tag['language_tag']:
            brand = lang_tag['value']
            break
    if 'model_name' in row:
        for lang_tag in (row['model_name']):
            if "en_" in lang_tag['language_tag']:
                model = lang_tag['value']
                break
    else:
        model = row['model_number'][0]['value']
        
    data_english_with_models.append({'title':title, 'brand': brand, 'model': model})
```

```python

df = pd.DataFrame.from_dict(data_english_with_models)

print(len(df))
```


- Remove rows without brand or model in title
- Convert encodings to characters
- Remove empty rowd

```python
def clean_data(df):
    special_chars = ['','-','_','#','@','$','=','+','*','£','%',',','.']
    is_branded = df.apply(lambda x : str(x['brand']) in x['title'],axis=1)
    has_model = df.apply(lambda x : str(x['model']) in x['title'],axis=1)
    is_brand_real = df.apply(lambda x: str(x.brand).isdigit() == False and str(x.brand) not in special_chars and str(x.brand) != "find.",axis=1)
    filled_df = df.fillna('')
    is_html = filled_df.title.str.contains('getTime')
    # Replace entity html codes : 
    to_replace = {
      '&quot;': '"',
      '&copy;': '',
      '&reg;' : '',
      '&amp;': '&'
    }
    ## Apply  filters
    filled_df = filled_df.replace(to_replace)
    df_clean = filled_df[~is_html][is_branded][is_brand_real][has_model]
    # Drop duplicates
    df_clean = df_clean.drop_duplicates(subset='title',keep='first')

    return df_clean
```

```python
cleaned_data = clean_data(df)
```
